```python
import tensorflow
from tensorflow import keras
from keras.preprocessing import image # like opencv helps in handling image data
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img=image.load_img("train/cat.jpeg",target_size=(148,210))
```

```python
img
```

```python
img.size
```

```python
type(img) # python image library object not numpy
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        shear_range=0.2,
        zoom_range=0.2,
        width_shift_range=0.2,
        height_shift_range=0.2,
        horizontal_flip=True,
        # vertical_flip=True do not apply as it can lead to upside down cat not required
        fill_mode='nearest')
```

```python
img=image.img_to_array(img) # converted to numpy array as pil type is not accepted
```

```python
img.shape
```

```python
img.size
```

```python
import numpy as np
img = np.expand_dims(img, axis=0) # add another dimension for batch
```

```python
img.shape
```

```python
i=0
for output in datagen.flow(img,batch_size=1,save_to_dir="train_augmented"):
    i=i+1
    if i==10:
        break
```
